```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(X):
  return (X + 3) ** 2
```

```python
def gradient(X):
  return 2 * (X + 3)
```

```python
x = 8
lr = 0.1
iterations = 100
tolerance = 0.000001
delta = 1
```

```python
x_points = []
while delta > tolerance and iterations >= 0:
  xnew = x - lr * gradient(x)
  x_points.append(xnew)
  delta = abs(xnew - x)
  x = xnew
  iterations -= 1
```

```python
plt.plot(np.linspace(-10, 10, 1000), f(np.linspace(-10, 10, 1000)), color='r')
plt.scatter(x_points, f(np.array(x_points)))
```
